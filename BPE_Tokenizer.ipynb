```python
from collections import Counter

# Convert each word into characters + end-of-word symbol
vocab = Counter()

for word in words:
    chars = " ".join(list(word)) + " </w>"
    vocab[chars] += 1

print("Initial Vocabulary:\n")

for k, v in vocab.items():
    print(f"{k} : {v}")
```

```python
from collections import defaultdict

def get_pair_frequencies(vocab):
    pairs = defaultdict(int)

    for word, freq in vocab.items():
        symbols = word.split()

        for i in range(len(symbols) - 1):
            pair = (symbols[i], symbols[i + 1])
            pairs[pair] += freq

    return pairs
```

```python
pairs = get_pair_frequencies(vocab)

print("Pair Frequencies:\n")

for pair, freq in sorted(pairs.items(), key=lambda x: x[1], reverse=True):
    print(pair, ":", freq)
```

```python
best_pair = max(pairs, key=pairs.get)

print("Most Frequent Pair:")
print(best_pair)
print("Frequency:", pairs[best_pair])
```

```python
import re

def merge_pair(pair, vocab):
    new_vocab = {}

    # Escape special characters for regex
    pattern = re.escape(" ".join(pair))
    pattern = re.compile(r'(?<!\S)' + pattern + r'(?!\S)')

    for word in vocab:
        # Replace the pair with the merged token
        new_word = pattern.sub("".join(pair), word)
        new_vocab[new_word] = vocab[word]

    return new_vocab
```

```python
vocab = merge_pair(best_pair, vocab)

print("Vocabulary After Merge:\n")

for word, freq in vocab.items():
    print(f"{word} : {freq}")
```

```python
num_merges = 20          
merge_rules = []

for i in range(num_merges):
    pairs = get_pair_frequencies(vocab)

    if not pairs:
        break

    best_pair = max(pairs, key=pairs.get)

    merge_rules.append(best_pair)

    vocab = merge_pair(best_pair, vocab)

    print(f"Merge {i+1}: {best_pair}")
```

```python
print("Learned Merge Rules:\n")

for i, rule in enumerate(merge_rules, start=1):
    print(f"{i}. {rule}")
```

```python
final_vocab = set()

for word in vocab:
    final_vocab.update(word.split())

print("Final Vocabulary:\n")

for token in sorted(final_vocab):
    print(token)

print("\nVocabulary Size:", len(final_vocab))
```

```python

token_to_id = {}
id_to_token = {}

for idx, token in enumerate(sorted(final_vocab)):
    token_to_id[token] = idx
    id_to_token[idx] = token

print("Token IDs:\n")

for token, idx in token_to_id.items():
    print(f"{idx} : {token}")
```

```python
def encode(word):
    # Split into characters
    tokens = list(word) + ["</w>"]

    # Apply each learned merge rule
    for pair in merge_rules:
        i = 0
        while i < len(tokens) - 1:
            if tokens[i] == pair[0] and tokens[i + 1] == pair[1]:
                tokens[i:i + 2] = ["".join(pair)]
            else:
                i += 1

    return tokens
```

```python
print(encode("tokenization"))
print(encode("attention"))
print(encode("machine"))
```

```python
def decode(tokens):
    word = "".join(tokens)
    return word.replace("</w>", "")
```

```python
tokens = encode("machine")

print("Encoded:", tokens)
print("Decoded:", decode(tokens))
```

```python
test_words = [
    "automation",
    "chatbot",
    "prediction",
    "language",
    "transformer"
]

for word in test_words:
    encoded = encode(word)
    decoded = decode(encoded)

    print("-" * 40)
    print("Original :", word)
    print("Encoded  :", encoded)
    print("Decoded  :", decoded)
```
